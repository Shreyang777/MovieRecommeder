```python
# convert str to list

import ast 
ast.literal_eval(movies.iloc[0].genres)
type(ast.literal_eval(movies.iloc[0].genres))
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])

    return L
```

#### handling genres

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

#### handling keywords

```python
# same for keywords
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

#### handling cast

```python
movies['cast'][0]
```

```python
### getting the name of top 3 actors
def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

#### handling crew

```python
movies['crew'][0]
```

```python
def director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(director)
```

#### handling overview

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())  ### to concatenate with others list
```

```python
movies.sample(5)
```

```python
## collapse all words to consider as a single enitity
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
###new_df will have 
### movie_id,title,tags(overview+genres+keywords+cast+crew)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.sample()
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_df.head(5)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
### converting to lower case to avoid any discripancies
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
new_df.shape
```

### stemming on tags 


```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem('dancing')
```

```python
def stem(text):
    y=[]

    for i in text.split(): # string to list 
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
stem(new_df['tags'][0])
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
        
```
